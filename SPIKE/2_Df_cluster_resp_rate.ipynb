# Quantification of neuron response rate

### Concaténation

```python
import pandas as pd
import os
import glob

# Définir le chemin du dossier contenant les fichiers Excel
folder_path = "U:/PhD/Experimentation/Electrophysiology/S1_recordings_Analysis/spyking_analysis/reponse_data"

# Trouver tous les fichiers correspondant au motif "Neuron_resp_*.xlsx"
files = glob.glob(os.path.join(folder_path, "Neuron_resp_*.xlsx"))

# Initialiser une liste pour stocker les DataFrames
dfs = []

# Parcourir chaque fichier et lire son contenu
for file in files:
    # Lire le fichier Excel dans un DataFrame
    df = pd.read_excel(file)
    
    # Ajouter le DataFrame à la liste
    dfs.append(df)

# Concaténer tous les DataFrames en un seul
concatenated_df = pd.concat(dfs, ignore_index=True)

# Afficher un échantillon des données concaténées pour vérifier
print(concatenated_df.head())

# Enregistrer le DataFrame concaténé dans un fichier Excel ou CSV si nécessaire
cluster_resp_path = "U:/PhD/Experimentation/Electrophysiology/S1_recordings_Analysis/spyking_analysis/cluster_response_rate"
output_file = os.path.join(cluster_resp_path, "grouped_neuron_resp.xlsx")
concatenated_df.to_excel(output_file, index=False)


```

# tri par conditions

```python
import pandas as pd
import os

# Charger le fichier trié
final_output_path = "U:/PhD/Experimentation/Electrophysiology/S1_recordings_Analysis/spyking_analysis/cluster_response_rate/grouped_neuron_resp.xlsx"
df = pd.read_excel(final_output_path)

# Créer une colonne 'condition' en combinant 'cdt', 'side' et 'genotype'
df['condition'] = df['cdt'] + '_' + df['side'] + '_' + df['genotype']

# Créer le dossier de sortie si il n'existe pas
output_folder = "U:/PhD/Experimentation/Electrophysiology/S1_recordings_Analysis/spyking_analysis/cluster_response_rate"
os.makedirs(output_folder, exist_ok=True)

# Liste des conditions uniques
unique_conditions = df['condition'].unique()

# Sauvegarder un fichier par condition unique
for condition in unique_conditions:
    # Filtrer les données pour cette condition
    df_condition = df[df['condition'] == condition]
    
    # Créer le chemin du fichier de sortie pour cette condition
    output_file_name = f"{condition}_resp_rate.xlsx"
    output_path = os.path.join(output_folder, output_file_name)
    
    # Sauvegarder le DataFrame filtré dans un fichier Excel
    df_condition.to_excel(output_path, index=False)
    
    print(f"Le fichier pour la condition '{condition}' a été sauvegardé à {output_path}")
```
