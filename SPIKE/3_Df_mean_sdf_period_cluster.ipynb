# Calculation of mean by cluster for each period

```python
import pandas as pd
import os

# Spécifier le répertoire contenant les fichiers à traiter
sdf_base_dir = "U:/PhD/Experimentation/Electrophysiology/S1_recordings_Analysis/spyking_analysis/sdf_data"

# Liste tous les fichiers dans le répertoire spécifié qui commencent par "SDF_downs"
files = [f for f in os.listdir(sdf_base_dir) if f.startswith("SDF_downs") and f.endswith(".xlsx")]

# Chemin du dossier de sortie
output_folder = "U:/PhD/Experimentation/Electrophysiology/S1_recordings_Analysis/spyking_analysis/mean_sdf_period_cluster"
os.makedirs(output_folder, exist_ok=True)

# Traitement de chaque fichier
for file in files:
    file_path = os.path.join(sdf_base_dir, file)
    
    # Chargement des données
    df_sdf = pd.read_excel(file_path)
    
    # Assurez-vous que la colonne 'sdf' est bien numérique
    df_sdf['sdf'] = pd.to_numeric(df_sdf['sdf'], errors='coerce')  # Remplace les valeurs non numériques par NaN
    
    # Convertir les colonnes en chaînes de caractères
    df_sdf['period'] = df_sdf['period'].astype(str)
    df_sdf['cdt'] = df_sdf['cdt'].astype(str)
    df_sdf['side'] = df_sdf['side'].astype(str)
    df_sdf['genotype'] = df_sdf['genotype'].astype(str)
    df_sdf['age'] = df_sdf['age'].astype(str)
    df_sdf['sex'] = df_sdf['sex'].astype(str)
    df_sdf['depth'] = df_sdf['depth'].astype(str)
    df_sdf['aCSF'] = df_sdf['aCSF'].astype(str)
    df_sdf['TTLtemp'] = df_sdf['TTLtemp'].astype(str)
    
    # Sélectionner les colonnes à conserver
    columns_to_keep = ["directory", "file_name", "neuron_id", "period", "cdt", "side", "genotype", "age", "sex", "depth", "aCSF", "TTLtemp", "sdf"]
    
    # Filtrer le DataFrame pour ne garder que les colonnes nécessaires
    df_sdf_filtered = df_sdf[columns_to_keep]
    
    # Grouper par les colonnes d'identification, et calculer la moyenne pour la colonne 'sdf' uniquement
    df_grouped = df_sdf_filtered.groupby(['directory', 'file_name', 'neuron_id', "cdt", "side", "genotype", "age", "sex", "depth", "aCSF", "TTLtemp", 'period'])['sdf'].mean().reset_index()
    
    # Créer le chemin du fichier de sortie pour ce fichier
    output_file_name = f"{os.path.splitext(file)[0]}_mean_sdf_period_cluster.xlsx"
    output_path = os.path.join(output_folder, output_file_name)
    
    # Sauvegarder le DataFrame groupé dans un fichier Excel
    df_grouped.to_excel(output_path, index=False)
    
    print(f"Le fichier a été sauvegardé à {output_path}")
```

# Concatenation en 1 fichier

```python
import pandas as pd
import os

# Spécifier le répertoire contenant les fichiers Excel générés
output_folder = "U:/PhD/Experimentation/Electrophysiology/S1_recordings_Analysis/spyking_analysis/mean_sdf_period_cluster"

# Liste tous les fichiers dans le répertoire spécifié qui commencent par "_grouped.xlsx"
files = [f for f in os.listdir(output_folder) if f.endswith("_mean_sdf_period_cluster.xlsx")]

# Liste pour stocker les DataFrames
all_data = []

# Traitement de chaque fichier généré
for file in files:
    file_path = os.path.join(output_folder, file)
    
    # Chargement des données
    df = pd.read_excel(file_path)
    
    # Ajouter le DataFrame à la liste
    all_data.append(df)

# Concaténer tous les DataFrames dans un seul DataFrame
final_df = pd.concat(all_data, ignore_index=True)

# Chemin du fichier de sortie pour le fichier concaténé
final_output_path = os.path.join(output_folder, "all_grouped_files_concatenated.xlsx")

# Sauvegarder le DataFrame concaténé dans un fichier Excel
final_df.to_excel(final_output_path, index=False)

print(f"Le fichier concaténé a été sauvegardé à {final_output_path}")
```
